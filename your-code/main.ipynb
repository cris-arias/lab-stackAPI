#### Question 1: Find the questions and answers of last month.

```python
#get data for questions

from stackapi import StackAPI
SITE = StackAPI('stackoverflow')

# convert to dataframe for easier reference by running json_normalize one time
# when running json_normalize for the first time, I get a one-single row df. 
# in order to disclose the info from the column, I need to run it again entering into the column that I want to disclose (in this case, 'items')

questions = SITE.fetch('questions', fromdate=1590962400, todate=1593468000)
questions
```

```python
#covert data to dataframe for easier reference
from pandas.io.json import json_normalize
questions_df = pd.json_normalize(questions['items'])
questions_df
```

```python
#get data for answers

answers = SITE.fetch('answers', fromdate=1590962400, todate=1593468000)
answers
```

```python
# convert to dataframe for easier reference by running json_normalize one time
# when running json_normalize for the first time, I get a one-single row df. 
# in order to disclose the info from the column, I need to run it again entering into the column that I want to disclose (in this case, 'items')

answers_df = pd.json_normalize(answers['items'])
answers_df
```
